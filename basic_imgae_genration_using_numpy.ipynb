```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = np.zeros((500,500))
image[:,:]= 100

plt.figure(figsize=(3,3))
plt.imshow(image,cmap= "gray")
plt.show()
```

```python
image[100:400,100:400]=255
image[200:300,200:300]=100
plt.figure(figsize=(3,3))
plt.imshow(image,cmap= "gray")
plt.show()
```
